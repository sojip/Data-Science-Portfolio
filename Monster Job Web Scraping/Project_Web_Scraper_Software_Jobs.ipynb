<h1>Project : Web scraper that fetches Software Developer job </h1>
    

In this project, we will build a web scraper that fetches Software Developer job listings from the <a href="https://www.monster.com">Monster job</a> aggregator site. 

```python
#import libraries and request the Monster Job site 
import requests #to send http request 
from bs4 import BeautifulSoup #to parse html content
import pandas as pd

URL = "https://www.monster.com/jobs/search/?q=Software-Developer&where=Africa"
page = requests.get(URL)

#create BeautifulSoup object
soup = BeautifulSoup(page.content, 'html.parser')

#get the HTML object that contains all of the search results
results = soup.find(id="SearchResults")
print(results.prettify)
```

Every job posting is wrapped in a <b>section</b> element with the class <b>card-content</b>. 

```python
job_elements = results.find_all('section', class_ = 'card-content')
if job_elements:
    print ("ok")
```

The page has descriptive class names on some elements. Let’s pick out only those:

```python
jobs = []
for job_element in job_elements:   
    job = {}
    job_title = job_element.find('h2', class_ = "title")
    job_company = job_element.find('div', class_ = "company")
    job_location = job_element.find('div', class_ = "location")
    if None in (job_title, job_company, job_location):
        continue
    job["title"] = job_title.text.strip()
    job["company"] = job_company.text.strip()
    job["location"] = job_location.text.strip()
    jobs.append(job)
print(jobs)
```

We can then save our resuts in a dataframe and in a csv file later.

```python
#save results in a dataframe
df = pd.DataFrame(jobs)
df.head()
```

```python
#save as csv file
df.to_csv('sofware_developer_jobs.csv', index = False)
```
